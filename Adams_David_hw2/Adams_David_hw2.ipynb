```python
# 1.i.

# Imports
import numpy
import matplotlib.pyplot as plt
import seaborn

# Allow notebook to display graph
%matplotlib inline

# Create time vector over 40 years with a step of .1 years (dt)
time_range = numpy.arange(0, 40, .1)

# Set up storage vectors and initial condition of 60000 otters
vec1 = numpy.zeros(len(time_range))
vec1[0] = 60000
vec2 = numpy.zeros(len(time_range))
vec2[0] = 60000

# Integrate using Euler's method using rate constant k = .5, carrying capacity K = 60000, rate of sea otters eaten = 
# 640 otters per whale per year, and number of whales = 5
for i in range(len(time_range) - 1):
    dN = (.5 * vec1[i] * (1 - (vec1[i] / 60000)) - (640 * 5)) * .1
    vec1[i + 1] = vec1[i] + dN

# Integrate using Euler's method using rate constant k = .5, carrying capacity K = 60000, rate of sea otters eaten = 
# 640 otters per whale per year, and number of whales = 10
for i in range(len(time_range) - 1):
    dN = (.5 * vec2[i] * (1 - (vec2[i] / 60000)) - (640 * 10)) * .1
    vec2[i + 1] = vec2[i] + dN

# Plot the graph
plt.plot(time_range, vec1, '-', label = '5 Whales')
plt.plot(time_range, vec2, '-', label = '10 Whales')
plt.xlabel('Time (years)')
plt.ylabel('Otters')
plt.legend()
```
