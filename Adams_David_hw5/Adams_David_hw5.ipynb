1.b.
    
    10^4 kW*hr * 1 yr / 365 days * 1 day / 24 hr = 1.1 kW
    * 1000 W / 1 kW = 1100 W
    
    1100 W / 1700 W/m^2 = 0.66 m^2
    
    An actual solar power would need to be larger than 0.66 m^2 because this estimate is assuming the solar
    panel is 100% efficient. Solar panels are nowhere near that efficient (most being between 10% and 20% efficient)
    and as such a larger solar panel is required. Also, solar panels can only generate electricity during daylight,
    which means that a surplus of power must be generated during the day and stored to be used at night.

2.a.

    * 1 kBT / 4.1*10^-21 J = 4 kBT
    
2.b.

    ΔGchem = -T(ΔSout + ΔSin)
    
2.c.

    W(N,Ω) = Ω choose N = (Ω*(Ω-1)*(Ω-2)*...*(Ω-N+1)) / N!
    since Ω >> N = (Ω*Ω*Ω*...*Ω) / N! = Ω^N / N!
    
2.d.

    ΔSout = Soutf - Souti = kB * ln(Wf) - kB * ln(Wi) = kB * ln(Wf/Wi)
    Wf = Ω^(N-1) / (N-1)!
    Wi = Ω^N / N!
    ΔSout = kB * ln((Ω^(N-1) / (N-1)!) / (Ω^N / N!)) = kB * ln(N/Ω)
    
    ΔSin = Sinf - Sini = kB * ln(Wf) - kB * ln(Wi) = kB * ln(Wf/Wi)
    Wf = Ω^(M+1) / (M+1)!
    Wi = Ω^M / M!
    ΔSin = kB * ln((Ω^(M+1) / (M+1)!) / (Ω^M / M!)) = kB * ln(Ω/(M+1))
    since M+1 = M
    = kB * ln(Ω/M)
    
2.e.
    
    ΔSout = kB * ln(N/Ω) = kB * ln(cout)
    ΔSin = kB * ln(Ω/M) = kB * ln(1/cin)
    ΔGchem = -T(ΔSout + ΔSin) = -T(kb * (ln(cout) + ln(cin))) = -kBT * ln(cout/cin) 
           = kBT * ln(cin/cout)
           
2.f.
    
    cin = 10^-pHin
    cout = 10^-pHout
    ΔGchem = kBT * ln(cin/cout) = kBT * ln(10^-pHin/10^-pHout) = kBT * ln(10^(pHout - pHin)
           = kBT * ΔpH * ln(10)
           
    ΔGtot = ΔGchem + ΔGelec = kBT * ΔpH * ln(10) + 4 kBT = (4 + ΔpH * ln(10)) kBT
    
2.g.

    ΔGchem = ΔpH * ln(10) kBT = 1.0 * ln(10) kBT = 2.3 kBT
    
    ΔGtot = (4 + ΔpH * ln(10)) kBT = 6.3 kBT
