```python
# 2.a.

# Imports
import Bio.SeqIO

def compare_sequences(seq1, seq2):
    """
    Compares 2 sequences of the same length and computes the percent similarity between them ignoring gaps.
    
    Parameters
    ----------
    seq1, seq2 : string
        DNA sequences, must be the same length
    
    Returns
    -------
    score : float
        percent similarity between seq1 and seq2
    """
    if len(seq1) != len(seq2):
        raise ValueError('Sequences must be of the same length')
        
    seq1 = seq1.lower()
    seq2 = seq2.lower()
    
    num_sim = 0
    num_comp = 0
    
    for base in range(len(seq1)):
        if seq1[base] != '-' and seq2[base] != '-':
            if seq1[base] == seq2[base]:
                num_sim += 1
            num_comp += 1
    
    score = num_sim / num_comp
    
    return score

# Open the files for the Sao Tome and Africa sequences
with open('hw1_dataset/frogs_st.txt', 'r') as seq:
    st_file = Bio.SeqIO.parse(seq, 'fasta')
    st_sequences = [record for record in st_file]
with open('hw1_dataset/frogs_africa.txt', 'r') as seq:
    africa_file = Bio.SeqIO.parse(seq, 'fasta')
    africa_sequences = [record for record in africa_file]

# Compare each of the 3 frogs from Sao Tome to the frogs from Africa and print top 3 results
c = 1
for place1 in st_sequences:
    # Initialize variables to hold best 3 matches and scores for them
    match1 = africa_sequences[0]
    match1_score = compare_sequences(place1.seq, match1.seq)
    match2 = africa_sequences[1]
    match2_score = compare_sequences(place1.seq, match2.seq)
    match3 = africa_sequences[2]
    match3_score = compare_sequences(place1.seq, match3.seq)
    
    # Sort our initial best 3 in order
    if(match2_score > match1_score):
        temp = match1
        temp_score = match1_score
        match1 = match2
        match1_score = match2_score
        match2 = temp
        match2 = temp_score
    if(match3_score > match1_score):
        temp = match1
        temp_score = match1_score
        match1 = match3
        match1_score = match3_score
        match3 = temp
        match3_score = temp_score
    if(match3_score > match2_score):
        temp = match2
        temp_score = match2_score
        match2 = match3
        match2_score = match3_score
        match3 = temp
        match3_score = temp_score
    
    # Compare sequences and find best 3 matches
    for x in range(3, len(africa_sequences)):
        score = compare_sequences(place1.seq, africa_sequences[x].seq)
        if(score > match1_score):
            match1 = africa_sequences[x]
            match1_score = score
        elif(score > match2_score):
            match2 = africa_sequences[x]
            match2_score = score
        elif(score > match3_score):
            match3 = africa_sequences[x]
            match3_score = score

    # Print stuff
    print('Comparing to frog ' + str(c) + ' from Sao Tome (' + place1.description + '):')
    print('Top 3 Matches:')
    print('1. ' + match1.description + ': Score = ' + str(match1_score))
    print('2. ' + match2.description + ': Score = ' + str(match2_score))
    print('3. ' + match3.description + ': Score = ' + str(match3_score))
    print()
    c += 1

# 2 of the frogs in the data set (which happen to be the 2nd and 3rd best matches) give the same score
# The 2nd and 3rd best matches are different frogs though
```
